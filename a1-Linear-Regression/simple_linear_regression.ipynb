```python
# Imports. 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generate random data points for Linear regression.

np.random.seed(101)
m = 180  
x = 3.7 * np.random.rand(m, 1)
y = 5.1 + 2.6 * x + np.random.randn(m, 1)

# Storing the values as tuples.
data_tuples = [(x[i][0], y[i][0]) for i in range(m)]
```

```python
# Plot the data.

plt.scatter(x, y, label="Data Points")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# Define a function for linear regression.

def SimpleRegression(data_tuples):
    """Compute a linear regression line.

    Parameters:
    ---
    data_tuples - list[tuples]
        Pairs of x and y data points to copmute the regression line with. 
    """
    x_mean = sum(x for x, _ in data_tuples) / len(data_tuples)
    y_mean = sum(y for _, y in data_tuples) / len(data_tuples)
    x_minus_xmean = [(x - x_mean) for x, _ in data_tuples]
        
    numerator = sum((x * y - x_mean * y_mean) for x, y in data_tuples)
    denominator = sum(x**2 for x in x_minus_xmean)
        
    slope = numerator / denominator
    intercept = y_mean - slope * x_mean
        
    return {"slope": slope, "intercept": intercept}
```

```python
# Run the regression function. 

regression = SimpleRegression(data_tuples)
result = SimpleRegression(data_tuples)
print(result)
```

```python
slope = result['slope']
intercept = result['intercept']
regression_line = [slope * x + intercept for x in x]

plt.scatter(x,y, label="Data Points")
plt.plot(x, regression_line, label="Regression Line", color="red")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend()
plt.show()
```
