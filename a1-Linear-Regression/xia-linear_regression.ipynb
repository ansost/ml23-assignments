```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import itertools
```

```python
np.random.seed(10)
m = 200  # number of instances

X = 3.2 * np.random.rand(m, 1)
y = 5.5 + 2.7 * X + np.random.randn(m, 1)
```

```python
# plot the randomly generated data
plt.scatter(X, y)
plt.plot(X, y, "b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.xlim(xmin=0)
plt.ylim(ymin=0)
plt.show()
```

```python
data = list(zip(list(itertools.chain.from_iterable(X)), list(itertools.chain.from_iterable(y))))
```

```python
class SimpleRegression():
    
    def __init__(self,data):
        self.data = data
        self.slope = None
        self.intercept = None


    def get_slope_intercept(self):
        X = [ item[0] for item in self.data ]
        y = [ item[1] for item in self.data ]

        X_mean = sum(X) / len(X)
        y_mean = sum(y) / len(y)

        X_minus_X_mean = [ n - X_mean for n in X ]
        slope_num = sum([ n*m for n,m in zip(X_minus_X_mean, y) ])
        slope_den = sum([ (n-X_mean)**2 for n in X ])

        self.slope = slope_num / slope_den

        self.intercept = y_mean - self.slope * X_mean
        
        return (self.slope, self.intercept)


    def plot_data(self):
        plt.scatter(X, y)
        plt.plot(X, y, "b.")
        plt.xlabel("$x_1$")
        plt.ylabel("$y$", rotation=0)
        plt.xlim(xmin=0)
        plt.ylim(ymin=0)
        plt.axline(slope=self.slope,xy1=(0.0,self.intercept))
        plt.show()
```

```python
sreg = SimpleRegression(data=data)

w0_w1 = sreg.get_slope_intercept()
```
