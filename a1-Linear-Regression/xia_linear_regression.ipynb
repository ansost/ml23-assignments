```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./house_price_data.csv")

X = list(df['area'])
y = list(df['price'])
```

```python
class LinearRegression():
    
    def __init__(self):
        self.slope = None
        self.intercept = None


    def get_slope_intercept(self,X, y):
        X_mean = sum(X) / len(X)
        y_mean = sum(y) / len(y)

        X_minus_X_mean = [ n - X_mean for n in X ]
        slope_num = sum([ n*m for n,m in zip(X_minus_X_mean, y) ])
        slope_den = sum([ (n-X_mean)**2 for n in X ])

        self.slope = slope_num / slope_den

        self.intercept = y_mean - self.slope * X_mean
        
        return True


    def plot_data(self):
        plt.scatter(list(df['area']), list(df['price']))
        plt.xlabel("area (square foot)")
        plt.ylabel("prices")
        plt.xlim(xmin=0)
        plt.ylim(ymin=0)
        plt.axline(slope=self.slope,xy1=(0.0,self.intercept))
        plt.show()
```

```python
lr = LinearRegression()

lr.get_slope_intercept(X, y)

print("weight (slope):   " + str(lr.slope) + "\n" +  "bias (intercept): " + str(lr.intercept))
```

```python
lr.plot_data()
```
