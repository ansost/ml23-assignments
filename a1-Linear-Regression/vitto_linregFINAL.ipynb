```python
#Creating a function

def SimpleRegression(data_tuples):
    x_mean = sum(x for x, _ in data_tuples) / len(data_tuples)
    y_mean = sum(y for _, y in data_tuples) / len(data_tuples)
    x_minus_xmean = [(x - x_mean) for x, _ in data_tuples]
        
    numerator = sum((x * y - x_mean * y_mean) for x, y in data_tuples)
    denominator = sum(x**2 for x in x_minus_xmean)
        
    slope = numerator / denominator
    intercept = y_mean - slope * x_mean
        
    return {"slope": slope, "intercept": intercept}

```
